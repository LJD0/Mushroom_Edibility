```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf
```

```python
# Read in the Mushroom Dataset from the csv 
# link to data: https://datahub.io/machine-learning/mushroom/r/1.html

mushroom_data_import_df = pd.read_csv('mushroom_csv.csv')
```

```python
# First dataframe transformations benchmark

mushroom_data_df = mushroom_data_import_df
```

```python
# View the DataFrame
mushroom_data_df
```

```python
# Clean column names

mushroom_data_df.rename(columns={'bruises%3F': 'bruises'},inplace=True)
```

```python
mushroom_data_df
```

### Encode data


```python
# Create an encoder

enc = OneHotEncoder(sparse=False)
```

```python
# Encode the Mushroom dataframe

mushroom_encoded_df = pd.DataFrame(enc.fit_transform(mushroom_data_df))
mushroom_encoded_df
```

```python
# Insert Column Names

mushroom_encoded_df.columns = enc.get_feature_names(mushroom_data_df.columns)
mushroom_encoded_df
```
