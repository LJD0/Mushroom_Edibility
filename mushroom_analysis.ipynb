```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf
import os
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
# Read in the Mushroom Dataset from the csv 
# link to data: https://datahub.io/machine-learning/mushroom/r/1.html

mushroom_data_import_df = pd.read_csv('Data/mushroom_csv.csv')
```

```python
# First dataframe transformations benchmark

mushroom_data_df = mushroom_data_import_df
```

```python
# View the DataFrame
mushroom_data_df
```

```python
# Clean column names

mushroom_data_df.rename(columns={'bruises%3F': 'bruises'},inplace=True)
```

```python
mushroom_data_df
```

### Encode data


```python
# Create an encoder

enc = OneHotEncoder(sparse=False)
```

```python
# remove 'class' column before encoding
pre_enc_df = mushroom_data_df.drop(columns='class')
```

```python
# Encode the Mushroom dataframe

mushroom_encoded_df = pd.DataFrame(enc.fit_transform(pre_enc_df))
mushroom_encoded_df
```

```python
# Insert Column Names

mushroom_encoded_df.columns = enc.get_feature_names(pre_enc_df.columns)
mushroom_encoded_df
```

```python
# Benchmark dataframe
# mushroom_encoded_df.to_csv('Data/mushroom_encoded.csv')
```

### Split Data for Testing and Training

```python
# Set up the y variable for training/testing

y = mushroom_data_df['class']
y=y.apply(lambda x: x.replace('e','1'))
y=y.apply(lambda x: x.replace('p','0'))
y=y.apply(lambda x: int(x))
```

```python
# set up X variable

X = mushroom_encoded_df
```

```python
# Split the dataset

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=52)
```

```python
# data does not need scaling
```

### Build and Compile the model

```python
# Define the model

input_feat = len(X_train.columns)
layer1_node = 30
layer2_node = 15

# initialize model
Mushroom_Model = tf.keras.models.Sequential()


# Add Hidden Layers

Mushroom_Model.add(
    tf.keras.layers.Dense(units=layer1_node,activation='relu',input_dim=input_feat)
)

Mushroom_Model.add(
    tf.keras.layers.Dense(units=layer2_node,activation='relu')
)

# Create an Output layer

Mushroom_Model.add(
    tf.keras.layers.Dense(units=1,activation='sigmoid')
)
```

```python
# Check model structure

Mushroom_Model.summary()
```

```python
# Compile the model
Mushroom_Model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

### Train and Checkpoint the Model

```python
# Define the checkpoint path and filenames
os.makedirs("Data/InitialRun/checkpoint/",exist_ok=True)
checkpoint_path = "Data/InitialRun/checkpoint/weights.{epoch:02d}.hdf5"

# create a callback that saves every five epochs

cp_callback = ModelCheckpoint(
    filepath=checkpoint_path, 
    verbose=1, 
    save_weights_only=True,
    save_freq='epoch',
    period=5)
```

```python
# Fit the model

fit_mushroom = Mushroom_Model.fit(
    X_train,
    y_train,
    epochs=50,
    callbacks=[cp_callback]
)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = Mushroom_Model.evaluate(X_test,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
